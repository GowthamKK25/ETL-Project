```python
import pandas as pd
import csv  
import datetime as dt
from sqlalchemy import create_engine
```

```python
dow_path = "dow_stock_info.csv"
nasdaq_path = "ndq_m_stock_info.csv"

dow_data = pd.read_csv(dow_path)
nasdaq_data= pd.read_csv(nasdaq_path)
```

```python
dow_df=pd.DataFrame(dow_data)
dow_df 
dow_df.dtypes 
```

```python
dow_df.drop(columns='Volume') 
```

```python
dow_df['Date'] = pd.to_datetime(dow_df['Date'], infer_datetime_format=True)  
```

```python
dow_df
```

```python
nasdaq_df=pd.DataFrame(nasdaq_data)
nasdaq_df 
```

```python
nasdaq_df.drop(columns="Volume") 
```

```python
nasdaq_df['Date'] = pd.to_datetime(nasdaq_df['Date'], infer_datetime_format=True) 
```

```python
nasdaq_df  
```

```python
merged_df=pd.merge(dow_df,nasdaq_df, how='left', on=["Date","Date"], suffixes=("_nasdaq","_dow"))
merged_df.dtypes
```

```python
# merge df=stock_db
rds_connection_string = "postgres:Postgres@localhost:5432/stock_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
merged_df.to_sql(name='stock_table', con=engine, if_exists='append', index=False)
```
